# 음료수 얼려 먹기 프로그램

N*M크기의 얼음틀이 주어지는 경우</br>
0은 얼음틀 구멍을 1은 칸막이를 표현</br>
0으로 연결된 묶음을 얼린 음료수 1개로 간주</br>

첫번째 입력은 N과 M, 1~1000 사이의 자연수</br>
두번째 입력은 얼음틀 형태가 주어짐</br>

탐색 알고리즘과 관련된 문제!

탐색 알고리즘

탐색이란</br>
많은 양의 데이터 중에서 원하는 데이터를 찾는 과정</br>
순차탐색, 이진 탐색, 해시탐색 등의 방법이 있음</br>
- 순차(Sequential)</br>
    왼쪽에서 오른쪽 방향으로 순서대로 하나씩 확인해 나가는 방법</br>
    주로 리스트 구조이며 데이터 수가 많아지면 시간이 많이 소요되어 효율이 떨어짐</br>
- 이진(Binary)</br>
    반 씩 쪼개가면서 찾는 방법</br>
    오름차순 또는 내림차순으로 정렬되어 있는 경우에 적합</br>
    평균적으로 순차보다 빠르지만, 데이터의 양, 구조, 정렬 상황 등에 따라 선택</br>
- 해시(Hash)
    데이터의 내용과 저장한 곳의 요소(인덱스)를 미리 연계하여 짧은 시간 안에 탐색</br>
    해시 함수를 이용해 데이터가 저장될 곳의 위치를 계산하고 해시함수에 의해 계산된 위치에 해시 테이블을 사용하여 데이터 저장 / 딕셔너리 구조와 비슷</br>

    데이터가 저장된 형태가 트리 또는 그래프인 경우</br>

    DFS(Deep First Search), 깊이 우선 탐색</br>
    BFS(Breadth First Search), 너비 우선 탐색</br>

    그래프와 트리의 공통점과 차이점</br>
    공통점 : 노드와 간선으로 구성 / 방향성 유무</br>
    차이점 : 그래프는 개체 간의 관계를 표현_네트워크, 최단거리/최소비용 문제 해결에 용이</br>
    트리는 관계와 더불어 계층을 표현</br>
    - 가중치가 없으며, 사이클이 존재하지 않음
    - 두 정점간 최단 경로는 단 하나만 존재
    - 간선의 개수는 정점개수 -1을 가짐



DFS 원리
"스택" 자료구조를 이용하여 탐색_push와 pop(first in last out)
- 탐색 시작 노드를 스택에 삽입하고 방문처리
- 스택의 최상단 노드에 방문하지 않은 인접 노드가 있으면 그 인접 노드를 스택에 저장
- 방문하지 않은 인접노드가 없을 시 스택의 최상단 노드를 제거
- 위의 과정을 모든 노드를 방문할 때까지 수행
- 인접노드가 여러 개일 경우 노드값 또는 가중치가 작은 순으로 선택

[1,2,7,6] 6 pop</br>
[1,2,7,8] 8 pop 7 pop 2 pop</br> 
[1,3,4,5] 5 pop 4 pop 3 pop 1 pop

노드 방문 순서(스택에 들어간 순서)
1 2 7 6 8 3 4 5

함수가 이런 형태를 띔 가장 먼저 호출된 함수가 가장 마지막에 메모리 해제됨</br>
연산자 처리

```
# 파이썬 코드로 구현해 봅시다
def dfs_f(data, visit_n):
    dfs_v[visit_n] = True   # 방문한 노드 번호에 True값을 넣어줌
    print(visit_n, end=' ') # 데이터를 한 줄에 출력해주기 위해서 end속성 사용
    for i in data[visit_n]:
        if not dfs_v[i]:    # false라면 그 다음 인자를 찾아가도록
            dfs_f(data, i)


data=[[],[2,3,8],[1,7],[1,4,5],[3,5],[3,4],[7],[2,6,8],[1,7],]
dfs_v=[False]*len(data) # 방문여부를 저장하는 1차원 리스트
print(dfs_f(data,1))

#질문1. 선생님 dfs와 bfs의 메인코드로 쓰였떤 'data'있잖아요. 어떻게 ppt의 그래프가 저렇게 표현되는거에요?
#1번노드가 2,3,8과 실선으로 연결되어 있습니다, 따라서 전체 데이터를 2차원 리스트의 형태로 저장 시 1번 노드와 연결되어 있는 [2,3,8] 을 1번 노드로 표현되는 1번째 인덱스에 입력한 것입니다

#질문2. data에서 0번째 인덱스의 값을 []로 넣어주는 이유는 뭔가요?
#노드 1을 2차원리스트의 인덱스 번호로 표현하다보니, 0번째 인덱스는 비워두고 1번째 인덱스부터 데이터를 입력한 것 입니다
#0번째 인덱스를 1번노드 표현으로 하고 싶으시면 코드상에서 +1 이라던가 하는 식의 이중 작업이 필요한 경우가 있어, 코드 상의 편리성으로 0번째 인덱스를 비웠습니다 
```

BFS원리
DFS가 가장 멀리 있는 노드를 우선으로 탐색하는데 비해 가까운 노드부터 탐색하는 알고리즘
큐 자료구조를 이용
- 탐색 시작 노드를 큐 구조에 삽입(선입선출)_프린터 출력 순서
- 큐에서 노드를 꺼내고 해당 노드의 인접 노드 중에서 방문하지 않은 노드를 모두 큐에 입력
- 바로 위의 과정 한 단계를 큐가 비워질 때까지 진행

[1] </br>
[2,3,8] 방문이 완료된 1은 먼저 빠져나감</br>
[3,8,7] 2가 나가면서 7이 들어옴</br>
[8,7,4,5] 3이 나가면서 4,5가 들어옴</br>
[7,4,5] 8번만 나감</br>
[4,5,6] 7이 나가면서 6이 들어오고 이후로는 모두 방문한 노드이기 때문에</br> 순차적으로 빠져나감</br>

노드 방문 순서(큐에서 빠져나온 순서)
1 2 3 8 7 4 5 6

```
from collections import deque

# 원래 큐는 선입선출 입구와 출구가 다름!
# deque는 출입구가 정해져 있지 않음 앞뒤로 드나들 수 있음

def bfs_f(data,start):
    queue = deque([start])
    bfs_visited[start] = True
    while queue:
        n = queue.popleft()
        print(n, end=' ')
        for i in data[n]:
            if not bfs_visited[i]:
                queue.append(i)
                bfs_visited[i] = True

data=[[],[2,3,8],[1,7],[1,4,5],[3,5],[3,4],[7],[2,6,8],[1,7],]
dfs_v=[False]*len(data) # 방문여부를 저장하는 1차원 리스트
print(dfs_f(data,1))
```
