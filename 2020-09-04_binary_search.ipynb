# 이진 탐색_binary search

1차원 데이터가 정렬되어 있을 때, 특정 데이터를 효율적으로 탐색하는 알고리즘</br>
데이터를 미리 정렬하여 최악의 경우에도 logN번의 항목 비교만 하는 효율적 탐색 방법
삽입 및 삭제가 빈번하게 발생할 시 정렬을 유지하기 위해 최악의 경우 O(n)이 소요되는 단점이 있음
차라리 해시 탐색을 사용하는 것이 효율적!

```
# 알고리즘

binary_search(left, right, k): # left 첫번째 인덱스 right: 마지막 인덱스 k:찾으려는 인덱스
    if left > right:
        return None
    middle = (left+right)//2
    if binary[middle] == k:
        return middle
    if binary[middle] > k:
        binary_search(left, middle-1, k)    # 중간 인덱스 앞부분 탐색
    else:
        binary_search(middle+1, right, k)   # 중간 인덱스 뒷부분 탐색
```

# 이진 탐색 트리(binary search tree)
이진 탐색 트리는 이진탐색의 개념을 트리형태의 구조에 접목한 자료구조</br>
이진탐색을 수행하기 위해 단순 연결 리스트를 변형</br>
이진 탐색 트리의 정의</br>
- 모든 노드의 키는 유일
- 왼쪽 서브 트리의 키들은 루트의 키보다 작음
- 오른쪽 서브 트리의 키들은 루트의 키보다 큼
- 즉, 루트 노드가 기준값이 되는 것!
- 왼쪽과 오른쪽 서브 트리도 이진 탐색 트리

탐색 작업을 효율적으로 하기 위한 자료구조</br>
key(왼쪽 서브 트리) <= key(루트 노드) <= key(오른쪽 서브 트리)</br>
이진 탐색을 중위순회 -> 오름차순 정렬</br>

뷸균형 트리가 되면 순차탐색과 별반 다르지 않음(데이터가 정렬되어 있는 상태)</br>
균형 잡힌 트리를 만들어주는 방법으로 AVL 트리 방식을 고안</br>
왼쪽 이진 트리와 오른쪽 이진 트리의 차수가 2 이상 차이나면 트리를 회전시켜 루트노드를 변경하고 트리를 재구성하는 방식

- 삭제하려는 노드가 말단 노드인 경우 : NULL값 삽입
- 삭제하려는 노드가 하나의 왼쪽, 또는 오른쪽 서브 트리 중 하나만 가지고 있는 경우
    - 삭제하고 자식 노드는 부모 노드에 연결
- 삭제하려는 노드가 두 개의 서브 트리를 모두 가지고 있는 경우
    - 삭제 노드와 가장 비슷한 값을 가진 노드를 서브 트리에서 찾아 삭제 노드의 위치에 연결
    - 다음의 노드 중에 임의로 하나 선택
    - 삭제 노드의 왼쪽 서브트리의 가장 큰 값인 맨 오른족 노드의 값
    - 삭제 노드의 오른쪽 서브트리의 가장 작은 값인 맨 왼쪽 노드의 값
